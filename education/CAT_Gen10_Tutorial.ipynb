## Introducció molt breu a l'execució de mètodes  
Com s'ha dit anteriorment, el paquet `gen10` et proporciona un conjunt de funcions relacionades amb la genòmica. Per executar aquestes funcions en Python primer has d'importar el paquet escrivint `import gen10`. Llavors pots utilitzar qualsevol mètode de `gen10` escrivint `sortida = gen10.nom_del_mètode(arguments)`. Anem pas a pas: utilitzarem el mètode `adn2arn()` com a exemple.

Primer, has de definir l'entrada que faràs servir com a argument del mètode. Un mètode és una funció del paquet `gen10` que pots cridar. El seu argument va dins dels parèntesis al final, i fa referència a l'entrada que necessita el mètode per funcionar. Si mires els detalls de `adn2arn()` a la taula anterior, veuràs que l'argument és una cadena de text (`string`) que conté una seqüència d'ADN. Pots definir-la així:
```python
adn = 'TACCACGTGGACTGAGGACTCCTCATT'
```
Un objecte `string` (`str`) en Python és una variable de text. S’escriu entre cometes simples `''` o dobles `""`.

Ara cridarem el mètode des de `gen10`. Per fer-ho escriu:
```python
gen10.adn2arn(adn)
```
Dins dels parèntesis posem el nom de la variable que volem utilitzar com a entrada. En cas que hi hagués més d’un argument, els hauries de separar amb una coma, com ara:
```python
gen10.simular_pcr(seq1, seq2, seq3)
```
Recorda que els arguments han de seguir estrictament l'estructura de la taula esmentada anteriorment, que també es troba al [README](https://github.com/joanalnu/gen10/blob/main/README.md#methods) del paquet.

Per guardar el resultat (sortida) de la funció executada, assignarem el seu valor a una variable:
```python
arn = gen10.adn2arn(adn)
```

Finalment, pots veure el resultat imprimint-lo amb el mètode natiu de Python `print()`:
```python
print(arn)
```

El codi complet és com la següent cel·la:

````python
# definir la secuenca d'ADN
adn = 'TACCACGTGGACTGAGGACTCCTCATT'
#adn = "TACCACGTGGACTGAGGACTCCTCATT" també es valid

# calcular l'equivalent d'ARN
arn = gen10.dna2rna(adn)
print(arn) # això imprimeix la variable arn
````

També podem emprar altres mètodes per calcular la seqüència aminoàcida d'aquesta proteïna.

````python
# calcular equivalent aminoàcid
aminoacids = gen10.rna2amino(arn)
print(aminoacids)

# també podem obtenir una versió compacta de la seqüència d'aminoàcids
aminoacids_compacte = gen10.asenzill(aminoacids)
print(aminoacids_compacte)
````

## Exemples d'Ús
A continuació trobarás alguns exemples usant mètodes de `Gen10`.

### 1. Creant una Mutació
Primer, generarem una mutació aleatoria en un seqüència d'ADN i la compararem amb la seqüència original.

````python
adn = 'TACCACGTGGACTGAGGACTCCTCATT'

# genera la mutació aleatoria
adn_mutat = gen10.crear_mutacio(adn)
print(adn_mutat)

# comparar per veure diferències
diferencia = gen10.compara(adn, adn_mutat)
print(diferencia)
````

També podem forçar una mutació en una posició específica and `mutar_lloc()`. En aquest cas utilitzarem una seqüència d'ARN.

````python
arn = gen10.adn2arn(adn) # obté l'ARN del nostre ADN de l'última cel·la

# defineix els paràmetres per a mutar_lloc()
posicio = 10
# aquesta no es la base que estem editant, sino un índex
# recorda que els ordinadors comencen comptant per 0, per tant la primera base té índex 0
# aquí estarem editant la base 11
base_nova = 'T'

# executem el mètode mutar_lloc
arn_mutat_al_lloc = gen10.mutar_lloc(arn, posicio, base_nova)

# verify mutation
diferencia_arn = gen10.compara(arn, arn_mutat_al_lloc)
````

### 2. Obrir un fitxer de text i utilitzar iteració  
`Gen10` té eines d'iteració integrades just on les necessites. Amb el mètode `iterar()` pots realitzar processos repetitius sobre una gran mostra de dades. Primer, comencem creant un fitxer de text pla (`.txt`) amb les nostres dades de mostra. Pots anar [aquí (G Drive)](https://drive.google.com/file/d/1nCGaT93zDf4n2K32yog5DcRBQTr9LiJZ/view?usp=sharing) per descarregar un fitxer d'exemple. Després, puja el fitxer a aquest notebook fent clic a la icona de la carpeta al menú de l'esquerra i seleccionant el teu fitxer.

````python
adns = gen10.llegir_input('example_file.txt') # aquest mètode llegueix l'input del nostre arxiu amb el nom example_file.txt
````

Ara definirem una llista (`list`) amb els nomes del mètodes que volem executar quan truquem `iterar` després.

````python
metodes = ['crear_mutacio', 'adn2arn', 'adn2amino']
sortida = gen10.iterar(adns, metodes) # executa el mètodes per totes les entrades
sortida # mostra la sortida que és una taula csv
````

### 4. Mètodes Simples
En aquesta última sub-secció et mostrem alguns exemples adicionals usant mètodes de `Gen10`.
