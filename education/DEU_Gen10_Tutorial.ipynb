## Sehr kurze Einführung zur Ausführung von Methoden

Wie oben erwähnt, stellt dir das `gen10`-Paket eine Reihe von Funktionen zur Verfügung, die sich auf Genomik beziehen. Um diese Funktionen in Python auszuführen, musst du zunächst das Paket importieren, indem du `import gen10` schreibst. Danach kannst du jede Methode von `gen10` mit folgendem Befehl verwenden: `output = gen10.method_name(arguments)`. Gehen wir Schritt für Schritt vor – wir verwenden die Methode `dna2rna()` als Demo.

Zuerst musst du die Eingabe definieren, die du als Argument für die Methode verwenden willst. Eine Methode ist eine Funktion des `gen10`-Pakets, die du aufrufen kannst. Das Argument kommt in die Klammern am Ende und beschreibt die Eingabe, die die Methode benötigt, um zu funktionieren. Wenn du dir die Details von `dna2rna()` in der obigen Tabelle ansiehst, wirst du feststellen, dass das Argument ein String ist, der eine DNA-Sequenz enthält. Du kannst ihn so definieren:
```python
dna = 'TACCACGTGGACTGAGGACTCCTCATT'
```
Ein String (`str`) in Python ist eine Textvariable. Er wird in einfache `''` oder doppelte `""` Anführungszeichen gesetzt.

Als Nächstes rufen wir die Methode aus dem `gen10`-Paket auf. Das machst du so:
```python
gen10.dna2rna(dna)
```
In die Klammern schreiben wir den Namen der Variable, die wir als Eingabe verwenden möchten. Falls mehrere Argumente notwendig sind, trennt man sie mit einem Komma, zum Beispiel:
```python
gen10.pcr_simulieren(seq1, seq2, seq3)
```
Denke daran, dass die Argumente strikt der Struktur in der oben erwähnten Tabelle folgen müssen – diese findest du auch im [README des Pakets](https://github.com/joanalnu/gen10/blob/main/README.md#methods).

Um das Ergebnis (Output) der ausgeführten Funktion zu speichern, weisen wir es einer Variable zu:
```python
rna = gen10.dna2rna(dna)
```

Zum Schluss kannst du dir das Ergebnis mit der Python-eigenen `print()`-Funktion anzeigen lassen:
```python
print(rna)
```

Der vollständige Code sieht dann so aus wie in der folgenden Zelle:

````python
# DNA-Sequenz definieren
dna = 'TACCACGTGGACTGAGGACTCCTCATT'
#  dna = "TACCACGTGGACTGAGGACTCCTCATT" ist auch ok

# ARN rechnen
rna = gen10.dna2rna(dna)
print(rna) # das drück die Variable rna aus
````

Jetzt können wir weitere Methoden einsetzen, um die Aminosäurenketten zu rechnen.

````python
# aminosäuren rechnen
aminosauren = gen10.rna2amino(rna)
print(aminosauren)

# wir konnen auch eine verkürzte Version haben
kompakte_aminosauren = gen10.zueinfach(aminosauren)
print(kompakte_aminosauren)
````

Es gibt viele unterschiedliche Methoden durch `Gen10` verfügbar, hier ist eine Kurzfassung von alle Methoden, ihre Funktion, Argumente und Outputs:

| 4 | rna2dna() | Transkribiert RNA-Strings zurück in DNA-Strings. | string | string |
| 5 | vergleichen() | Vergleicht die Strings (unabhängig davon, ob DNA, RNA oder Aminosäuren), gibt immer einen Boolean und einen String zurück. True, wenn beide Strings identisch sind, oder False und wo die Strings sich unterscheiden. | string1, string2 | boolean, string |
| 6 | checken() | Prüft, ob der bereitgestellte String eine gültige DNA- oder RNA-Sequenz ist. Prüft nicht auf Aminosäureketten. | string | string |

| 8 | mutation_erstellen() | Gibt einen neuen String mit einer Mutation zurück (nur 1 pro Ausführung). Die Mutation kann eine Base ändern, eine Base löschen oder eine neue an einer beliebigen Position hinzufügen. | string | string |
| 9 | iterieren() | Durch Eingabe einer Liste von Eingaben und einer Liste von Funktionen gibt es eine Tabelle mit allen Ergebnissen für jede Funktion und Eingabe zurück. | list, list | dataframe (Tabelle) |
| 10 | zueinfach() | Transkribiert eine Aminosäurekette vom Drei-Buchstaben-Code in den Ein-Buchstaben-Code. | string | string |
| 11 | alphafold_struktur() | Durch Eingabe einer UniProt-ID $^1$ gibt es eine URL zur `pdb`-Datei der vorhergesagten Proteinstruktur zurück. | string | dictionary |
| 12 | protein_generieren() | Durch Eingabe des resultierenden Wörterbuchs von `alphafold_struktur()` gibt es eine Visualisierung der vorhergesagten Proteinstruktur zurück. | dictionary | None |
| 13 | dna_scheiden(string, integer) | Schneidet den DNA-String an der angegebenen Position in zwei Teile. | string und integer | string Original-DNA mit markiertem Schnitt |
| 14 | dna_reparieren(string, string, integer, string) | Repariert einen geschnittenen DNA-String, indem entweder eine Base gelöscht wird (NHEJ) oder spezifische Basen an der angegebenen Stelle hinzugefügt werden (HDR). | string DNA-String, string Reparaturtyp (NHEJ oder HDR), integer Optional: Schnittposition, string Optional: einzufügender String bei HDR-Reparatur | string Reparierte DNA |
| 15 | finden(string, sequence) | Findet eine lokale Sequenz in einer größeren, globalen Sequenz. | string, string (global, lokal) | [(int, int)] Indizes der gefundenen Position |
| 16 | codon_checken(string) | Prüft auf nicht existierende Codons in einer DNA- oder RNA-Sequenz. | string | ['ABC'] Liste nicht existierender Codons |

| 18 | gc_gehalt(dna) | Berechnet den GC-Gehalt (Prozentsatz) einer gegebenen DNA-Sequenz. | string | float |
| 19 | schmelz_temperatur(dna) | Berechnet die Schmelztemperatur (Tm) einer kurzen DNA-Sequenz mit der Wallace-Regel. | string | float |

### 1. Erzeugen einer Mutation  
Zuerst erstellen wir eine zufällige Mutation in einer DNA-Sequenz und vergleichen sie mit der Originalsequenz.

````python
dna = 'TACCACGTGGACTGAGGACTCCTCATT'

# Erstellung eine zufällige Mutation
mutierte_dna = gen10.mutation_erstellen(dna)
print(mutierte_dna)

# Unterschieden suchen
unterschiede = gen10.compare(dna, mutierte_dna)
print(unterschiede)
````

Du kannst auch eine Mutation an einer bestimmten Stelle erzwingen, indem du `mutate_site()` verwendest. Jetzt werden wir eine RNA-Sequenz mutieren.

````python
rna = gen10.dna2rna(dna)  # RNA-Sequenz aus der DNA-Sequenz der vorherigen Zelle erzeugen

# Parameter für mutate_site() definieren
position = 10
# Das ist nicht die Base, die wir ändern, sondern der Index
# (Denk dran: Computer beginnen bei 0 zu zählen, d. h. die erste Base (Base 1) hat den Index 0)
neue_base = 'T'

# Die Methode mutate_site() aufrufen
mutierte_rna = gen10.stelle_mutieren(rna, position, neue_base)

# Mutation überprüfen
rna_unterschiede = gen10.vergleichen(rna, mutierte_rna)
````

Du kannst den Wert einer Variablen erhalten, indem du einfach ihren Namen ausführst:

````python
rna_unterschiede
````
